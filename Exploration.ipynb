```python
plt.figure(figsize=(12, 8))
plt.hist(df.EXC_RATE);
plt.title("Exchange rate from January to December 2019")
plt.xlabel("Exchange Rate")
plt.ylabel("Frequency");
```

### From the visualization above the frequency is tinted(skwed to the right) so there is need for more visual exploration to confirm this is true.

```python
count, bin_edges = np.histogram(df['EXC_RATE'], 10)
df['EXC_RATE'].plot(kind='hist', figsize=(8,5), xticks=bin_edges, bins=15);
plt.xlabel('Exchange Rate')
plt.show()
```

### Due to the fact that the viz still not showing much, lets create bins from the data point min and max

### **Rubric** Tips Visualizations should depict the data appropriately so that the plots are easily interpretable. You should choose an appropriate plot type, data encodings, and formatting as needed. The formatting may include setting/adding the title, labels, legend, and comments. Also, do not overplot or incorrectly plot ordinal data.

```python
binsize = 8
bins = np.arange(0, df['EXC_RATE'].max()+binsize, binsize)
count, bin_edges = np.histogram(df['EXC_RATE'])
df['EXC_RATE'].plot(kind='hist', figsize=(10,6), xticks=bin_edges,
                   bins=bins);
plt.xlabel('Exchange Rate')
plt.show()
```

## From the visualization above we can now confirm that exchange rate is skwed to the left due to the fact that we have being able to plot three different plot and we have same shape of graph

```python
# Plotting the histogram for fob value for different currencies
plt.figure(figsize=(10,6))
# plt.subplot(1, 2, 1)
# binsize = 1000000
count, bin_edges = np.histogram(df['FOB_VAL'])
df['FOB_VAL'].plot(kind='hist', xticks=bin_edges,
                   bins=10);
# plt.xscale('log')
plt.xlabel('FOB value in different currencies')
plt.title("Currency Value");
```

### The graph above shows that the frequency of free on board is skwed to the right, but there is the need to verify this because there seem to be high spread between the data points. So in this case am going to use a subplot to plot with and without log graph to show a clear difference and give a clear visualization of frequency. 

```python
plt.figure(figsize=(12,8))
plt.subplot(1, 2, 1)
count_1, bin_edges_1 = np.histogram(df['FOB_Value_Naira']) # the numpy histogram function is used to get the bin edges and count for each bin intervals
df['FOB_Value_Naira'].plot(kind='hist', xticks=bin_edges_1);
plt.xlabel('FOB_Value_Naira')
plt.title("Fob distribution without log")
plt.subplot(1, 2, 2)
log_binsize = 0.030
bins = 10 ** np.arange(2.5, np.log10(df['FOB_Value_Naira'].max())+log_binsize, log_binsize) # logging the max data point and adding the log binsize
# count, bin_edges = np.histogram(df['FOB_Value_Naira'])
df['FOB_Value_Naira'].plot(kind='hist',
                   bins=bins)
plt.title("Fob distribution with log");
plt.xscale('log')
plt.show()
```

### After verifying the fob frequency we can see from the above subplot that our fob value is uniformly distributed on a logarithm scale with the graph on the right.

## Bivariate Exploration

```python
plt.figure(figsize=(12,9))
plt.subplot(1, 2, 1)
sns.scatterplot(x= "EXC_RATE", y='FOB_Value_Naira', data=df);
plt.xlabel('Exchange Rate')
plt.ylabel('Free on Board Value')
plt.title("fob and Exch_rate without log")
plt.subplot(1, 2, 2)
sns.scatterplot(x= "EXC_RATE", y='FOB_Value_Naira', data=df);
plt.xlabel('Exchange Rate')
plt.yscale('log')
plt.ylabel('Free on Board Value')
plt.title("fob and Exch_rate with log")
plt.suptitle("Relationship btw FOB and Exchange Rate");
```

- The subplot with title without log between FOB and Exch_Rate shows the relationship between exchange rate and free on board value, but the relationship is not that clear so there is the need to take the log of the y scale. Thus, from the visualisation it is clear that there is relationship between the two numerical variables

```python
sns.set_theme(style="darkgrid")
g = sns.catplot(x='PAYMENT_MODE', y='CFR', data=df, 
kind='point', height=8)
plt.title("Count of payment mode");
```

- The graph above show that the average mode of payment for cost of freigt using bills for collection as a mode of payment is between 300000 and 350000 and the average cost of freight for letter of Credit is between 600000 and 650000 a significant difference right? which goes to tell us reason why importer tend to use more of Bills of Collection compare to Letters of Credits.

```python
sns.catplot(x='CURRENCY', data=df, hue='PAYMENT_MODE', kind='count', height=10);
plt.title("Dominant currency used for Transaction in 2019");
```

- From the graph above we see that the most predominant currency used for transaction in Nigeria for both Bill of Collection and Letters of Credit is the Dollar.

```python
sns.catplot(data=df, x='CURRENCY', y='EXC_RATE', kind='point', height=8)
sns.catplot(x='BNK_NAM', y="CFR", data=df, kind='point', join=False, height=8)
plt.xticks(rotation=90)
plt.title("Exchange rate of currency used for import to Naira")
plt.yscale('log');
```

- The first graph shows the currencies per exchange rate this is used to explore the exchange rate for all currencies in the distribution. 
- The second graph shows banks and cost of freight with and exponential scale, which clearly shows that union bank has the highest cost of freight.

```python
currency = df.groupby('CURRENCY', axis=0).sum()
```

```python
colors_list = ['red', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink', 'beige', 'aqua', 'cornflowerblue']
explode_list = [0.1, 0.5, 0, 0.2, 0.1, 0, 0, 0.1, 0.15] # ratio for each currency with which to offset each wedge.

currency.CFR.plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.0f%%', 
                            startangle=90,          
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            explode=explode_list # 'explode' lowest 3 currencies
                            )

# scale the title up by 12% to match pctdistance
plt.title('Proportion of cost of freight per currency', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=currency.index, loc='upper right');
```

 - From the visualization we can see that most freight transaction was done using the USD with 77% of import transaction done in 2019
