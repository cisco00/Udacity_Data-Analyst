### Mode of Payment

```python
def chart(dataframe, X, rot=0):
    sns.countplot(data=dataframe, x=X)
    sns.set_style('whitegrid')
    plt.xticks(rotation=rot)
```

```python
plt.figure(figsize=(10, 6))
chart(df, "PAYMENT_MODE")
plt.title("Distribution of Mode of Payment");
```

#### From the above graph it shows that Bill of Collection is the most used by importers of good into the country in 2019

### Banks that where used for transaction in 2019 

```python
plt.figure(figsize=(10, 8))
chart(df, "BNK_NAM", 90)
plt.title("Transaction rate per Bank", size = 20);
```

#### From the visualization above it shows that Zenith bank has the highest rate of transaction when it comes to import in 2019. But given that we have two forms of payment by the importer there is the need to investigate further to ascertain if actual Zenith bank can stand out from both form of transactions as the most used or otherwise.

### Banks that where used for Bills of Collections transaction in 2019 

```python
bank_payment_mode = df.loc[df.PAYMENT_MODE == "Bills for Collection"]
```

```python
plt.figure(figsize=(12, 10))
bank_payment_mode.BNK_NAM.value_counts().plot()
plt.title("Banks that recieve Bill of Collection as mode of payment", fontsize = 20);
```

- From the above visualization it is clear that the Zenith bank was the most used for Bills for Collections transaction in 2019 

```python
bank_payment_mode1 = df.loc[df.PAYMENT_MODE == "Letters of Credit"]
```

```python
plt.figure(figsize=(12, 10))
bank_payment_mode1.BNK_NAM.value_counts().plot()
plt.title("Banks that recieve Letters of Credit as mode of payment", fontsize = 20);
```

#### Further investigation into banks with high transaction rate for import in 2019 tells us that more people tend to use zenith bank for import transaction. 

### Currency used for import transaction in nigeria in 2019 

```python
plt.figure(figsize=[12, 8])
sns.catplot(x='CURRENCY', data=df, hue='PAYMENT_MODE', kind='count')
plt.title("Dominant currency used for Transaction in 2019", size = 1);
```

#### From the above visualization it shows the currency that was used by importers in nigeria most was the USD with over 70% transactions.

## Proportion of cost of freight per currency

```python
currency = df.groupby('CURRENCY', axis=0).sum()
```

```python
colors_list = ['red', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink', 'beige', 'aqua', 'cornflowerblue']
explode_list = [0.1, 0.5, 0, 0.2, 0.1, 0, 0, 0.1, 0.15] # ratio for each currency with which to offset each wedge.

currency.CFR.plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.0f%%', 
                            startangle=90,          
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            explode=explode_list # 'explode' lowest 3 currencies
                            )

# scale the title up by 12% to match pctdistance
plt.title('Proportion of cost of freight per currency', y=1.12, size = 20) 

plt.axis('equal') 

# add legend
plt.legend(labels=currency.index, loc='upper right');
```

 - From the visualization we can see that most freight transaction was done using the USD with 77% of import transaction done in 2019

### Checking for relationship between variables

```python
plt.figure(figsize=(12, 10))
sns.set_theme(style="ticks")
sns.relplot(x='FREIGHT_VAL', y='FOB_VAL', data=df,
           kind='scatter', hue='CURRENCY', size='CURRENCY',
          alpha=0.6, palette='inferno', style='CURRENCY')
plt.xscale('log')
plt.yscale('log')
plt.title('log Relationship between Free on board value and freight value per currency', fontsize = 20);
```

- From the above visualization it shows that there is high correlation between FOB and FREIGHT_val, and the most dominant currency used when it comes to  transaction in freight and free on board is basically dollars, then euros while others are sometimes rarely used for such transaction..

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='CFR', y='FOB_VAL', data=df, color='seagreen')
plt.xscale('log')
plt.yscale('log')
plt.xticks([1e2, 1e3, 1e4, 1e5, 1e6, 1e7, 1e8], ['100', '1k', '10k', '100k', '1m', '10m', '100m'], size=12)
plt.yticks([1e1, 1e3, 1e5, 1e7], ['10', '1k', '100k', '10m'], size=12)
plt.xlabel('Cost of Freight', fontname='Times New Roman')
plt.ylabel('Free On Board Value', fontname='Times New Roman')
plt.title('Relationship between Cost of freight and Free on Board Value', size=20, fontname='Times New Roman');
```

#### The above visualization shows a clear and a precise relationship between Cost of Freight and Free on Board
